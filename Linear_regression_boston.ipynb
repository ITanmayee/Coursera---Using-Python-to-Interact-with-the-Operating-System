```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/Boston.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
features = data.drop(['medv'], axis=1)
target = data['medv']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
print(X_train.shape)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
print(lr_model.intercept_, lr_model.coef_)
```

```python
train_pred = lr_model.predict(X_train)
test_pred = lr_model.predict(X_test)

print(f"Training error: {sqrt(mean_squared_error(y_train, train_pred)):.3f}")
print(f"Testing error: {sqrt(mean_squared_error(y_test, test_pred)):.3f}")
```
